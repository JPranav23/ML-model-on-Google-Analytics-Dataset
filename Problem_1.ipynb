```python
# Importing necessary libraries

import numpy as np
import pandas as pd
import seaborn as sn
import array
import random
import sklearn
from sklearn.tree import export_graphviz
from IPython.display import Image
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Importing data

sample_user_data = pd.read_csv("~/Desktop/Board Infinity/sample_user_data.csv", low_memory=False)
sample_user_data.info()
sample_user_data.head()
print(sample_user_data.columns)
len(sample_user_data.index)

# Checking missing values

sample_user_data.isnull().sum()

# We see that total_transactions column has a lot of missing values.
# We cannot simply replace missing values by mean/median.
# Therefore we need to delete those rows from our dataset.

# Handling missing values

sample_user_data = sample_user_data[sample_user_data['totals_transactions'].notna()]

# Checking the missing values now
sample_user_data.isnull().sum()
len(sample_user_data.index)

# As we see that totals_bounces is a complete column of missing values after 1st round of cleaning data.
# Therefore, we need to delete that column and move ahead with making predictive models.

sample_user_data = sample_user_data.drop("totals_bounces", axis=1)

# Again checking null values
sample_user_data.isnull().sum()

# As we see, "totals_timeOnSite" column also has 1 missing values.
# So we could drop that row.

sample_user_data = sample_user_data[sample_user_data['totals_timeOnSite'].notna()]

# Again checking null values
sample_user_data.isnull().sum()
len(sample_user_data.index)

# We have 6307 rows with no missing values.

# Now let us divide our 'data_dropped_rows' data into training and testing datasets

X = sample_user_data.drop(['fullVisitorId', 'date', 'totals_totalTransactionRevenue', 'totals_transactions', 'visitStartTime', 'device_mobileDeviceModel'], axis=1)
X = pd.DataFrame(X)
Y = sample_user_data['totals_transactions']
Y = pd.DataFrame(Y)
Y['totals_transactions'] = np.where(Y['totals_transactions']>=1, 'Yes', 'No')

X.info()

# Factorizing datatype
for x in X:
    if X[x].dtypes == "object":
        X[x],class_names = pd.factorize(X[x])

X.info()

Y.totals_transactions,class_names = pd.factorize(Y.totals_transactions)

# Splitting the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, test_size=0.3, random_state=0)

# Building the decision tree model

clf = DecisionTreeClassifier()
clf.fit(X_train, Y_train)

# Predicting from model

Y_pred = clf.predict(X_test)
Y_pred

# Confusion matrix and classification report

print(metrics.confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))

# As we can see, the decision tree model has an accuracy of 1 which clearly means that this is an over-fitted model.
```
